```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("telco_rawdata.csv")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Offer']=data['Offer'].fillna("No Offer")
data['Internet Type']=data['Internet Type'].fillna("None")
data['Churn Category']=data['Churn Category'].fillna("No Churn")
data['Churn Reason']=data['Churn Reason'].fillna("No churn")
```

```python
data.isnull().sum().sum()
```

```python
data.describe()
```

```python
data.columns
```

```python
import seaborn as sns
```

```python
churn_counts = data['Churn Label'].value_counts()
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%')
plt.title("Churn Distribution")
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn Label', data=data, palette='Set1')
plt.title("Churn by Contract Type")
plt.show()
```

```python
sns.boxplot(x='Churn Label', y='Monthly Charge', data=data, palette='coolwarm')
plt.title("Monthly Charges vs Churn")
plt.show()

```

```python
sns.countplot(x='Paperless Billing', hue='Churn Label', data=data, palette='Set2')
plt.title("Churn by Paperless Billing Status")
```

```python
sns.histplot(data=data, x='Tenure in Months', hue='Churn Label', multiple='stack', bins=30, palette='Set2')
plt.title("Tenure Distribution by Churn")
plt.show()
```

```python
selected_cols = [
    'Age','Number of Referrals','Tenure in Months','Monthly Charge',
    'Total Charges','Total Revenue','Satisfaction Score','Churn Score','CLTV'
]
numeric_data = data[selected_cols]

plt.figure(figsize=(12,8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (Key Features)")
plt.show()

```

```python
sns.countplot(x='Payment Method', hue='Churn Label', data=data)
plt.title("Churn by Payment Method")
plt.xticks(rotation=45)  # rotate labels for readability
plt.show()
```

```python
sns.countplot(x='Internet Type', hue='Churn Label', data=data)
plt.title("Churn by Internet Service Type")
plt.show()
```

```python
sns.boxplot(x='Churn Label', y='Satisfaction Score', data=data)
plt.title("Satisfaction Score vs Churn")
plt.show()
```

```python
data['tenure_group'] = pd.cut(data['Tenure in Months'], bins=[0,12,36,72],labels=['New','Mid','Loyal'])

sns.countplot(x='tenure_group', hue='Churn Label', data=data)
plt.title("Churn by Tenure Group")
plt.show()
```
